```python
def read_stock_csv(stock_file):
    df = pd.read_csv(
        f"./data/{stock_file}",
        index_col="Date",
        parse_dates=True,
        infer_datetime_format=True)
    try:
        return df["Volume"].astype("float")
    except:
        return df["Volume"]
```
